# Veri Çerçevesi ile grafik oluşturma

Şimdi de bir veri çerçevesi oluşturmalıyız. Böylece çizimlerimizde bazı güzel eğilimleri görebileceğiz. Bu kodumuzda rastgele sayılar üreteceğiz ve görselleştireceğiz.

Rastgele sayı üretmek için NumPy'ın random kütüpanesinden .random() fonksiyonunu kullanacağız. Sayıların tutarlı olarak üretilmesi için .seed() komutunu kullanıyoruz.

```python
np.random.seed(1221)
# Rastgele sayı üretmek için NumPy'ın random kütüpanesinden .random() fonksiyonunu kullanacağız.
# Sayıların tutarlı olarak üretilmesi için .seed() komutunu kullanıyoruz.
```

```python
df = pd.DataFrame()
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2
# Boş bir veri çerçevesi oluşturup, İçini 0 ve 1 arasında rastgele değerlerle dolduralım.
# BU sayılarına iki fazlasını ve karesini de yeni kolonlar olarak veri çerçevemize ekleyelim.
```

```python
df['counts_sq'] = df.index ** 2 
df['counts_sqrt'] = np.sqrt(df.index)
# Son olarak veri çerçevemizin indexini kullanarak iki yeni kolon daha oluşturalım.
```

```python
plt.plot(df['rand'])
plt.show()
```

# 2. Gösterim seçenekleri

```python
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1, 1.1])
plt.ylabel('Değerler')
plt.title('Rastgele Sayılar')
plt.show()
# Yukarıda oluşturduğumuz çizge, gayet basit bir grafik, 
# ancak grafiklerimizi daha belirgin ve bilgilendirici hale getirmek için yapabileceğiniz birçok şey var.
# Diyelim ki rengi mor olarak değiştirmek,y-eksenini genişletmek,y-eksenini etiketlemek ve çizgeye bir başlık vermek istiyoruz.
```

```python
plt.plot(df['rand'], color='purple')
plt.plot(df['rand_shift'], color='green')
plt.ylim([-0.1, 3.1])
plt.ylabel('Değerler')
plt.title('Rastgele Diziler')
plt.show()
```

# 5. Birden fazla grafik çizdirmek

 Matplotlib'in tanıtmak istediğimiz bir işlevselliği daha var: subplot. Subplot, matplotlib’in tek bir figürde birden çok plot çizge yoludur. Figürünüzü çok farklı alanlara ayırıyorsunuz ve her alanda farklı çizgeler oluşturuyorsunuz. Bu, bir dizi grafik oluşturmak ve bunları açıkça ilişkilendirilmiş bir şekilde sunmak için son derece yararlı olabilir. Aynı zamanda birden çok görselin tek bir resim olarak aktarılmasını da kolaylaştırabilir. Aşağıdaki görsellerin tek bir resim olduğuna dikkat edin.

Yukarıdaki iki çizgeyi bir figürde iki subplot olarak bir araya getirelim. Bunun için matplotlib'in subplot işlevselliğini kullanacağız. Bunu yapmak için, her çizge oluşturmadan önce bir subplot tanımlamamız gerekecek. Bu, üç argüman alan plt.subplot() kullanılarak yapılır. İlk iki parametre figürün boyutlarını tanımlar, üçüncü ise o noktada oluşturduğunuz subplotu tanımlar.

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Değerler')
plt.title('Rastgele Diziler')

plt.subplot(1, 2, 2)
plt.plot(df['rand_shift'], color='green')
plt.ylabel('Ötelenmiş Değerler')
plt.title('Ötelenmiş Diziler')
plt.show()
```

Böylece, solda rastgele değerlere sahip olan dizi ve sağda bunun ötelenmiş haliyle 1 yüksekliğinde ve 2 genişliğinde bir subplot oluşturduk. Güzel! Ama gerçekten nasıl çalıştı?

Evet, plt.figure() fonksiyonu, oluşturmak istediğimiz şeklin boyutunu, istediğimiz x ve y boyutlarıyla bir ikili (tuple) alarak ayarlar. Sonra her çizge plt.subplot() fonksiyonu kullanılarak oluşturulur. Bu, üzerinde çalışacağımız çizgeleri belirtir. Diğer bir deyişle, plt.subplot() fonksiyonu ilk iki parametreyle tüm figürün kaç satır ve sütundan oluştuğunu, üçüncü parametre ile de o anki çizgenin bu figürün kaçıncı çizgesi olacağını belirler.

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Değerler')
plt.title('Rastgele Diziler')

plt.subplot(1, 2, 2)
plt.scatter(x = df['rand_sq'], y = df['rand'], color='green')
plt.ylabel('Kare Değerler')
plt.title('Kare Diziler')

plt.tight_layout()
plt.show()
```

# 6. Histogram grafiği

Histogram, bir değişkenin olası değerlerini ve bu değerlerin ne kadar yaygın olduğunu göstermek için harika bir araçtır.
Şimdi bir tane çizdirelim. Öncelikle normal bir dağılımdan bazı rastgele değişkenler oluşturalım (normal dağılımın ne olduğunu bilmiyorsanız endişelenmeyin, sonraki bölümde öğreneceksiniz) ve 1000 tane rastgele sayı oluşturup bir histogram çizdirelim.

```python
x = np.random.normal(10, 5, 1000)

plt.hist(x) 
plt.show()
# random.normal() fonks. verdiğimiz 10 ve 5 sayıları, kullandığımız normal dağılımın merkezi ve standart sapmasını belirler.
```

Madem bir histogramımız var, şimdi nasıl çalıştığına dair biraz daha ayrıntıya girelim. Histogram, değerleri kutulara (bin) bölerek başlar. Bu kutular, üst ve alt sınırı olan eşit büyüklükteki aralıklardır. Histogram daha sonra her kutuda kaç tane değer bulunduğunu sayar ve her kutu için bu sayıya karşılık gelen bir çubuk çizer. Bu, değişkeninin aralığının yoğunluğu hakkında size bir fikir verir. Histogramlar, bir değişkenin dağılımını görselleştirmek için harikadır.

# 6.1. Kutu sayımı ve yerleştirme
Varsayılan olarak, matplotlib tüm değerleri on kutuya dağıtır ve bu kutular için en iyi yerleşimi seçer. Bu, istediğiniz sayı olmayabilir fakat kutu sayısı istediğiniz gibi ayarlanabilir

```python
x = np.random.normal(10, 5, 1000)

plt.hist(x, bins=40, color='red')
plt.title('Kutu Seçimi')
plt.xlabel('Rastgele Değerler')
plt.show()
# 1000 tane rastgele sayı oluşturup bir histogram çizdirelim fakat bu sefer rengi kırmızı, kutu sayısını 40 yapalım.
```

```python
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)

plt.hist(x, color='blue', bins=np.arange(-10, 40), alpha=.5) 
plt.hist(y, color='red', bins=np.arange(-10, 40), alpha=.5)
plt.title('Manuel Kutu Yerleşimi')
plt.xlabel('Rastgele Değerler')

plt.show()
```

# 6.2. Histogramların normalleştirilmesi
İki farklı değişkenin dağılımlarını yukarıdaki iki histogramla karşılaştırmanın zor olduğunu farkettiniz mi? Her değişkenin ölçekleri ve yükseklikleri farklı olduğundan kırmızı histogram görsele hakim oldu. Normalleştirme, bu sorunu, her bir histogram altındaki toplam alan 1 olacak şekilde kutuların yüksekliğini yeniden ölçeklendirerek çözmemize olanak tanır.

Bu, farklı ölçeklerde değişkenlere sahip olduğunuz ve bunların histogramlarını karşılaştırmak istediğimizde kullanışlıdır. Aşağıda yukarıdaki aynı verilerle fakat bu sefer normalleştirilmiş bir histogram çizimi görüyoruz:

```python
plt.hist(x, normed=True, color='blue', bins=np.arange(-10, 40), alpha=.5) 
plt.hist(y, normed=True, color='red', bins=np.arange(-10, 40), alpha=.5)
plt.title('Normalize Edilmiş Histogramlar')
plt.xlabel('Rastgele Değerler')

plt.show()
```

# 7. Boxplot (kutu) grafiği
Bir başka yararlı görselleştirme, boxplot yani kutu çizgesidir. plt.boxplot() fonksiyonu ile bir boxplot oluşturabilirsiniz. Boxplot nispeten basit bir görselleştirmedir. Şimdi bir tane oluşturalım ve daha sonra özelliklerini inceleyelim. Tek ihtiyacımız olan bir dizi veya değişken. Yine normal dağılımı kullanacağız. Rastgele veri oluşturup boxplot çizdirelim.

```python
np.random.seed(111)

x = np.random.normal(10, 5, 1000)

plt.boxplot(x)
plt.show()
```

Boxplot, çeyrekler açıklığı (interquartile range ya da kısaca IQR) denilen şeyi kapsayan bir kutudan oluşur. İlk olarak çeyrekler açıklığının ne olduğunu açıklayalım. Bir değişkeni, içerdiği değerlere göre aşağıdaki gibi dört çeyrek bölüme ayırabilirsiniz:
- En düşük değerden %25’e,
- %25’den medyana,
- Medyandan %75’e ve
- %75’den maksimum değere

Bu gruplara çeyrekler (quartiles) denir ve ortadaki ikisinin birlikte kapsadığı değerlere çeyrekler arası (interquartile range) denilir. Böylece boxplot'ta gördüğümüz kutu, %25’den %75’e kadar tüm verileri içermekte ve medyanı kırmızı bir çizgi ile göstermektedir.

Kutunun üstünde ve altında bıyıklar (whiskers) denilen dikey çizgiler bulunur (bu sebepten, boxplota bazen "kutu ve bıyık çizgeleri" de denilir). Bıyıklar, IQR kutusunun üstünden ve altından 1.5x uzaklığa kadar aşağı ve yukarı uzanacaktır. Tüm değerler bu aralıkta ise, bıyıklar değişkenin maksimum ve minimum değerlerine kadar gidecektir. IQR'ın 1.5 katının ötesindeki herhangi bir değer, bıyıkların sonunun ötesinde küçük bir kene (tick) olarak temsil edilecektir. Örneğin, yukarıdaki çizgede, bıyıkların üstünde ve altında birkaç değer var.

Boxplot, bir değişkenin yoğunluğu veya dağılımı hakkında bilgi edinmek için yararlı değildir. Bununla birlikte, değer aralığını araştırmak ve olası aykırı değerleri (outlier) tespit etmek için çok yararlıdır. Bir kutu çizgesi üzerinde bir nokta ya da daire ile işaretlenmiş veri noktaları genellikle aykırı değerler olarak kabul edilir. Bu nedenle, yeni bir veri üzerinde çalışırken kutu çizgeleri yaygın olarak kullanılır.
